```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "https://www.bikewale.com/best-bikes-in-india/"
```

```python
response = requests.get(url)
response.status_code
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content,"html.parser")#we can use lxml in place of html.parse
```

```python
print(soup.text)
```

```python
image = soup.findAll("div",class_="imageWrapper")
for i in image:
    j=i.img["src"]
    print(j)
    
```

```python
image
```

```python
img1=[]
image = soup.findAll("div",class_="imageWrapper")
for i in image:
    j=i.img["src"]
    print(j)
    img1.append(j)
print(img1)
```

```python
links =[]
link = soup.findAll("div",class_="bikeDescWrapper")
for i in link:
    j=i.a["href"]
    links.append(j)
links
```

```python
links =[]
link = soup.findAll("div",class_="bikeDescWrapper")
for i in link:
    j=i.a.text
    links.append(j)
links
```

```python
with open("bikewala.csv","w") as csv_file:
    write = csv.writer(csv_file)
    write.writerow(image)
    for i in image:
        j=i.img["src"]
        img1.append(j)
    write.writerow(img1)
```
